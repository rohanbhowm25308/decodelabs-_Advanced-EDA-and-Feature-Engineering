```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.boxplot(x=df["TotalPrice"])
plt.show()
```

```python
Q1 = df["TotalPrice"].quantile(0.25)
Q3 = df["TotalPrice"].quantile(0.75)

IQR = Q3 - Q1

lower = Q1 - 1.5 * IQR
upper = Q3 + 1.5 * IQR

df = df[(df["TotalPrice"] >= lower) &
        (df["TotalPrice"] <= upper)]
```

```python
sns.histplot(df["TotalPrice"], bins=30)
plt.show()
```

```python
sns.countplot(x="PaymentMethod", data=df)
plt.xticks(rotation=45)
plt.show()
```

```python
sns.countplot(x="OrderStatus", data=df)
plt.xticks(rotation=45)
plt.show()
```

```python
sns.countplot(x="ReferralSource", data=df)
plt.xticks(rotation=45)
plt.show()
```

```python
df["AvgItemPrice"] = df["TotalPrice"] / df["Quantity"]
```

```python
df["CartValue"] = df["ItemsInCart"] * df["UnitPrice"]
```

```python
df["CouponUsed"] = df["CouponCode"].apply(
    lambda x: 0 if x == "No Coupon" else 1
)
```

```python
df["Date"] = pd.to_datetime(df["Date"])

df["OrderMonth"] = df["Date"].dt.month
```

```python
numeric_cols = df.select_dtypes(include='number')

sns.heatmap(numeric_cols.corr(),
            annot=True,
            cmap="coolwarm")

plt.show()
```
